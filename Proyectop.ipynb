 ### <font color=green>PORTAFOLIO1 - 2016</font>

```python
def get_adj_closes(tickers, start_date=None, end_date=None, freq='m'):
    closes = web.YahooDailyReader(symbols=tickers, start=start_date, end=end_date, interval=freq).read()['Adj Close']
    closes.sort_index(inplace=True)
    return closes
```

```python
tickers = ['AMX','T','VZ','TMUS','DIS','JPM','WFC','V','HSBC','MET','AAPL','GOOG','FB','MSFT',
           'INTC','WMT','AMZN','COST','NKE','KO']
```

```python
closes1 = get_adj_closes(tickers,
                        start_date = '2010-01-01',
                        end_date='2015-12-31')
```

```python
closes1.plot(figsize=(13,7))
plt.grid()
plt.legend(bbox_to_anchor = (1,1))
```

#### Hacemos zoom en la gráfica para observar mejor los activos que cuentan un precio entre 0-150

```python
closes1.plot(figsize=(13,7))
plt.grid()
plt.legend(bbox_to_anchor = (1,1))
plt.ylim([0,150])
```

```python
ret1=(closes1.pct_change()).dropna()
```

```python
Eret1 = pd.DataFrame(data={'Media':ret1.mean()*12,
                                'Volatilidad':ret1.std()*np.sqrt(12)},
                           index=ret1.columns)
```

```python
cov1=ret1.cov()
corr1=ret1.corr()
```

```python
Eret1=Eret1.T
#Tasa libre de riesgo certificados de tesorería a un año reportados en la página de Departamento de tesorería de los EEUU el
#dia 16 de diciembre de 2015
rf = 0.0305

# Matriz cov
S= np.diag(Eret1.loc['Volatilidad'].values)
Sigma=S.dot(corr1).dot(S)

#Rendimientos esperados activos
Eind=Eret1.loc['Media'].values
```

```python
def var(w, Sigma):
    return w.dot(Sigma).dot(w)
def menos_RS(w,Eind,rf,Sigma):
    E_port=Eind.T.dot(w)
    s_port=var(w,Sigma)**0.5
    RS=(E_port-rf)/s_port
    return -RS
```

```python
N=len(Eind)
#Dato inicial
w0= np.ones(N)/N
# Cotas de las variables
bnds=((0,1), )*N
# Restricciones
cons = {'type':'eq', 'fun':lambda w: w.sum() - 1}
```

```python
emv=minimize(fun=menos_RS,
            x0=w0,
            args=(Eind,rf,Sigma),
            bounds=bnds,
            constraints=cons)
```

```python
w_emv=emv.x
E_emv=Eind.T.dot(w_emv)
s_emv=var(w_emv,Sigma)**0.5
RS_emv=(E_emv-rf)/s_emv
```

```python
g=10
w_opt=(E_emv-rf)/(g*s_emv**2)

Poemv=pd.DataFrame(data={'Pesos%':w_emv*100},index=ret1.columns)
Poemv=Poemv.round(4)
Poemv[Poemv['Pesos%']!=0]
```

```python
w_opt*w_emv,1-w_opt
```

Tomando en cuenta que si en verdad estuvieramos invirtiendo, a la hora de vender nuestros activos tendríamos que pagar el 10% sobre las utilidades, por lo que hemos decidido calcular el valor de nuestra inversión descontando este impuesto en cada año.

```python
impuesto2016 = ((inv2016 - inv_ini)*.1)
inv2016_new = inv2016 - impuesto2016
print('La cantidad de impuesto a pagar de este año es de:',round(impuesto2016,2),
      'resultadon con monto final de:',round(inv2016_new,2))
```

```python
inv2018_new = (1+r_total_2018/100)*inv2017_new
impuesto2018 = ((inv2018_new - inv2017_new)*.1)
inv2018_new = inv2018_new - impuesto2018
print('La cantidad de impuesto a pagar de este año es de:',round(impuesto2018,2),
      'resultadon con monto final de:',round(inv2018_new,2))
```

```python
inv2019_new = (1+r_total_2019/100)*inv2018_new
impuesto2019 = ((inv2019_new - inv2018_new)*.1)
inv2019_new = inv2019_new - impuesto2019
print('La cantidad de impuesto a pagar de este año es de:',round(impuesto2019,2),
      'resultadon con monto final de:',round(inv2019_new,2))
```

```python
inv2020_new = (1+r_total_2020/100)*inv2019_new
impuesto2020 = ((inv2020_new - inv2019_new)*.1)
inv2020_new = inv2020_new - impuesto2020
print('La cantidad de impuesto a pagar de este año es de:',round(impuesto2020,2),
      'resultadon con monto final de:',round(inv2020_new,2))
```

```python
inv2021_new = (1+r_total_2021/100)*inv2020_new
impuesto2021 = ((inv2021_new - inv2020_new)*.1)
inv2021_new = inv2021_new - impuesto2021
print('La cantidad de impuesto a pagar de este año es de:',round(impuesto2021,2),
      'resultadon con monto final de:',round(inv2021_new,2))
```
